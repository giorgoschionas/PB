```python
import numpy as np
import random 


grades = []
max_points = [10,10,10,10,10,10]
low = 2
for _, max_p in zip(range(len(max_points)),max_points):
    test =[]
    for _ in range(20):
        test.append(random.randrange(low,max_p+1))
    grades.append(test)
    low = low+1
    
grades_arr = np.array(grades)
grades_arr = np.transpose(grades_arr)
print(grades_arr)

courses = ['Μαθηματικά', 'Φυσική', 'Έκθεση', 'Αγγλικά', 'Λογοτεχνία', 'Γυμναστική']


with open('grades.csv', 'w', newline='') as myfile:
        wr = csv.writer(myfile)
        wr.writerow(courses)
        wr.writerows(grades_arr)
```
