```python
import numpy as np
import random 


grades = []
max_points = [10,10,10,10,10,10]
low = 2
for _, max_p in zip(range(len(max_points)),max_points):
    test =[]
    for _ in range(20):
        test.append(random.randrange(low,max_p+1))
    grades.append(test)
    low = low+1
    
grades_arr = np.array(grades)
grades_arr = np.transpose(grades_arr)
print(grades_arr)

courses = ['Μαθηματικά', 'Φυσική', 'Έκθεση', 'Αγγλικά', 'Λογοτεχνία', 'Γυμναστική']


with open('grades.csv', 'w', newline='') as myfile:
        wr = csv.writer(myfile)
        wr.writerow(courses)
        wr.writerows(grades_arr)
```

```python
import numpy as np
import names,random
from faker import Faker
from datetime import date
import csv

fake = Faker(0)


names_list =[]
for _ in range(12):
    names_list.append(names.get_full_name().split())

cryptos = ['BTC','ETH','ADA','SOL','AVAX','XTZ']
start_date, end_date = date(2022,3,20), date(2022,5,20)

transactions= []

# Txs: Tx ID, Tx date, Crypto, To, From, Amount 
i=0
while i<100:
    sender = ' '.join(random.choice(names_list))
    receiver = ' '.join(random.choice(names_list))
    if sender == receiver:
        continue
    tx_hash=random.randrange(10000,9999999)
    tx_date = fake.date_between(start_date='-1y', end_date='today')
    tx_asset = random.choice(cryptos)
    amount = random.randrange(1,1000)
    tx = [tx_hash,tx_date.strftime('%m/%d/%Y'),tx_asset,sender,receiver,amount]
    transactions.append(tx)
    i=i+1


print(transactions)

# transactions_arr = np.arr(tr)
    
    
with open('transactions.csv','w', newline='') as myfile:
    wr = csv.writer(myfile)
    wr.writerows(transactions)
```
